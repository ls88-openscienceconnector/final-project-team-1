We wish to analyze the relationship between two variables. In this study, we believed that conservative voters engaged with fake news more within the context of this study. WE create a data visualization that examines this relationship graphically.

$H_0$: Voters during the 2016 election are most likely to engage with fake news sources were conservative leaning, older, and highly engaged with political news

$H_A$: All types of voters during the 2016 election are engaging with fake news at similar frequencies. 

```python
domain_codings = pd.read_csv('Domain Coding Page Four.csv')
display(domain_codings.head(20))
```

```python
domain_codings.groupby('partisanship')['weighted political alignment'].plot.bar()
plt.ylabel('weighted political alignment')
plt.xlabel('partisanship')
```

Average weighted political alignment of websites that recieved exposure during 2016 election season:

```python
domain_codings['weighted political alignment'].mean()
```

### Conclusion:
The plot above looks at the relationship between partisanship and the weighted political alignment of website that voters visiting during the 2016 election season. The plot shows that more websites with higher weighted political alignment ratings, recieved more exposure than websites with lower weighted political alignment ratings. High ratings indicates a higer association to conservative ideals while lower rating indicate a higher association to libera ideals. The amount of websites visited with positive values of y are in higher density compared to negative values of y. Therefore, a higher percentage of conservative sites recieved exposure and interaction with voters. Moreover, the average weighted political alignment of websites that recieved exposure during 2016 election season also indicates that a higher percentage of conservative sites recieved exposure and interaction with voters than did sites promoting liberal ideals.
